```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_entrena_escalado = scaler.fit_transform(X_entrena)
X_prueba_escalado = scaler.fit_transform(X_prueba)

pd.DataFrame(X_entrena_escalado)
```

```
from sklearn.model_selection import cross_val_score

# Parámetros para validación cruzada
cv = 10  # Número de particiones
scoring = 'neg_mean_squared_error'  # Métrica: MSE negativo
```

```
# Regresión Lineal
from sklearn.linear_model import LinearRegression

modelo_lineal = LinearRegression()

# Realizar validación cruzada
scores_lineal = cross_val_score(modelo_lineal, X_entrena_escalado, Y_entrena, cv=cv, scoring=scoring)

# Convertir los scores a valores positivos y calcular métricas
mse_lineal_cv = -scores_lineal  # Negar para obtener el MSE
promedio_lineal_cv = mse_lineal_cv.mean()

# Imprimir resultados
print("Validación Cruzada - Regresión Lineal:")
print(f"Errores cuadráticos medios (MSE): {mse_lineal_cv}")
print(f"MSE Promedio: {promedio_lineal_cv}")
```

```
# Árboles de Decisión
from sklearn.tree import DecisionTreeRegressor

modelo_arbol = DecisionTreeRegressor(random_state=42)

# Realizar validación cruzada
scores_arbol = cross_val_score(modelo_arbol, X_entrena_escalado, Y_entrena, cv=cv, scoring=scoring)

# Convertir los scores a valores positivos y calcular métricas
mse_arbol_cv = -scores_arbol  # Negar para obtener el MSE
promedio_arbol_cv = mse_arbol_cv.mean()

# Imprimir resultados
print("Validación Cruzada - Árboles de Decisión:")
print(f"Errores cuadráticos medios (MSE): {mse_arbol_cv}")
print(f"MSE Promedio: {promedio_arbol_cv}")
```

```
# Random Forest
from sklearn.ensemble import RandomForestRegressor

modelo_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Realizar validación cruzada
scores_rf = cross_val_score(modelo_rf, X_entrena_escalado, Y_entrena, cv=cv, scoring=scoring)

# Convertir los scores a valores positivos y calcular métricas
mse_rf_cv = -scores_rf  # Negar para obtener el MSE
promedio_rf_cv = mse_rf_cv.mean()

# Imprimir resultados
print("Validación Cruzada - Random Forest:")
print(f"Errores cuadráticos medios (MSE): {mse_rf_cv}")
print(f"MSE Promedio: {promedio_rf_cv}")
```

```
# ---- Definición de Promedios y Desviaciones Estándar ----
promedio_lineal_cv = -scores_lineal.mean()  # Cambia a tu variable de resultados de regresión lineal
std_lineal_cv = scores_lineal.std()

promedio_arbol_cv = -scores_arbol.mean()  # Cambia a tu variable de resultados de Árbol de decisión
std_arbol_cv = scores_arbol.std()

promedio_rf_cv = -scores_rf.mean()  # Cambia a tu variable de resultados de Random Forest
std_rf_cv = scores_rf.std()
```

```
# ---- DATOS DE EVALUACIÓN ----
list_model = ['Linear Regression', 'DecisionTree', 'Random Forest']
list_mean = [promedio_lineal_cv, promedio_arbol_cv, promedio_rf_cv]  # Promedios de MSE de validación cruzada
list_sd = [std_lineal_cv, std_arbol_cv, std_rf_cv]  # Desviaciones estándar de MSE

# ---- GRÁFICA ----
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
for i in range(3):
    plt.errorbar(i, list_mean[i], xerr=0, yerr=list_sd[i], fmt='o', linewidth=3, label=list_model[i])

plt.xticks([0, 1, 2], list_model)
plt.title("Evaluación de Modelos (MSE y Desviación Estándar)")
plt.ylim(0, max(list_mean) + 2 * max(list_sd))  # Ajuste dinámico del límite superior del eje Y
plt.grid(visible=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(visible=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.ylabel("Mean Squared Error (MSE)")
plt.legend(title="Modelos")
plt.show()
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Definir el modelo de RandomForest
forest_reg = RandomForestRegressor()

# Definimos los posibles parámetros
n_estimators = [3, 10, 30, 50, 80, 100]
min_samples_split = [2, 4, 6, 8]
max_depth = [2, 4, 6, 8, 10]

# Definir los parámetros en un diccionario para GridSearchCV
params = {
    'n_estimators': n_estimators,
    'min_samples_split': min_samples_split,
    'max_depth': max_depth
}

# Aplicamos el GridSearchCV
grid_search = GridSearchCV(estimator=forest_reg, param_grid=params, cv=10, scoring='neg_mean_squared_error')

# Ajustar el modelo
grid_search.fit(X_entrena, Y_entrena)

# Mostrar los mejores parámetros encontrados
print("Mejores parámetros encontrados: ", grid_search.best_params_)

# Obtener el mejor modelo
best_forest_reg = grid_search.best_estimator_

# Calcular el score del mejor modelo en los datos de prueba
score_rf_best = best_forest_reg.score(X_prueba, Y_prueba)
print(f"Score del mejor modelo (Random Forest): {score_rf_best:.4f}")
```
